```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ------------------------------
# Load dataset
# ------------------------------
df = pd.read_csv("dataset.csv", skiprows=4)

# Filter for India
india_data = df[df["Country Name"] == "India"]

# Select latest year (2022)
population_2022 = india_data["2022"].values[0]
print("India Population 2022:", population_2022)

# ------------------------------
# Age Distribution Data (from UN / chart reference)
# ------------------------------
age_groups = ["0-20 Years", "21-64 Years", "65+ Years"]
population_values = [512, 807, 98]  # in millions
percentages = [36.1, 57.0, 6.9]

# ------------------------------
# Bar Chart
# ------------------------------
plt.figure(figsize=(8,6))
bars = plt.bar(age_groups, population_values, color=["gold", "royalblue", "crimson"])

# Add labels on top of each bar
for bar, pop, pct in zip(bars, population_values, percentages):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
             f"{pop}M\n({pct}%)", ha='center', fontsize=10, fontweight='bold')

plt.title("India's Population Distribution by Age (2022)", fontsize=14, fontweight='bold')
plt.ylabel("Population (in Millions)")
plt.xlabel("Age Groups")
plt.ylim(0, max(population_values) + 100)

# Save and Show
plt.savefig("population_bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()

# ------------------------------
# Histogram (simulated ages for visualization)
# ------------------------------
ages = np.concatenate([
    np.random.randint(0, 21, size=512000),
    np.random.randint(21, 65, size=807000),
    np.random.randint(65, 100, size=98000)
])

plt.figure(figsize=(8,6))
plt.hist(ages, bins=20, color="teal", edgecolor="black")
plt.title("India's Age Distribution Histogram (Simulated, 2022)", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Frequency")

# Save and Show
plt.savefig("population_histogram.png", dpi=300, bbox_inches="tight")
plt.show()
```
